### Условие 3: 3 задача
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0 # если малая корреляция, т.е. < 0.3, то приравниваем это 0 и не рассматриваем
corr_matrix
```

###### Сформирована матрица взаимосвязи (корреляции) факторов

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidth=1)
```

##### Сформирована тепловая диаграмма (heatmap) корреляции факторов
Например, цена объекта недвижимости зависит от жилой площади (коэф. 0,7). Также непосредственно взаимосвязаны - конструктив здания и цена (коэф. 0,7), конструктив здания и количество ванных комнат (коэф. 0,7), конструктив здания и жилая площадь (коэф. 0,8).

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['waterfront'], y=df['price'])
plt.xlabel('Вид на набережную')
plt.ylabel('Цена')
plt.title('Зависимость стоимости недвижимости от вида на набережную');
```

##### Выводы по диаграмме "Зависимость стоимости недвижимости от вида на набережную":
- стоимость объектов недвижимости БЕЗ ВИДА на набережную составляет в среднем 0,5 млн. При этом диапазон стоимости колеблется преимущественно от 80 тыс. до 1,0-1,1 млн. Хотя единичные объекты могут оцениваться достаточно высоко.
